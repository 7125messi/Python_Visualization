# 3 圆形颜色系统

当您有六个以上的类别需要区分时，最简单的方法是在圆形颜色空间中绘制均匀间隔的颜色（例如色调会保持亮度和饱和度不变）。

这是大多数seaborn函数在需要使用比当前在默认颜色循环中设置更多颜色时默认使用的功能。

最常见的方法是使用**hls颜色空间**，这是RGB值的简单转换。

```python
sns.palplot(sns.color_palette("hls", 8))
```

```python
sns.palplot(sns.color_palette("hls", 10))
```

```python
plt.figure(figsize=[8,6])
data = np.random.normal(size=(20,8)) + np.arange(8)/2
sns.boxplot(data=data,palette=sns.color_palette("hls", 8))
```

### hls_palette()功能可以让您控制颜色的亮度(lightness)和饱和度(saturation)

```python
sns.palplot(sns.hls_palette(8, l=.3, s=.8))
```

### husl可以很容易地选择均匀间隔的色调，同时保持明亮度和饱和度更加均匀

```python
sns.palplot(sns.color_palette("husl", 8))
```

```python
sns.palplot(sns.color_palette("hls", 8))
```

# 5 使用xkcd颜色来命名颜色

xkcd很多包来命名随机的RGB颜色，可以产生一组954种指定颜色，你现在可以使用xkcd_rgb字典在seaborn中引用

```python
plt.plot([0, 1], [0, 1], sns.xkcd_rgb["pale red"], lw=3)
plt.plot([0, 1], [0, 2], sns.xkcd_rgb["medium green"], lw=3)
plt.plot([0, 1], [0, 3], sns.xkcd_rgb["denim blue"], lw=3);
```

```python
colors = ["windows blue", "amber", "greyish", "faded green", "dusty purple"]
sns.palplot(sns.xkcd_palette(colors))
```
