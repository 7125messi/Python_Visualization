# 8 seaborn图形显示效果

## 8.1 axes_style and set_style

```python
x = np.linspace(0,14,100)
y1 = np.sin(x)
y2 = np.sin(x+2)*1.25
```

```python
def sinplot():
    plt.plot(x,y1)
    plt.plot(x,y2)
```

```python
sinplot() # 应用了seaborn的画图风格
```

```python
style = ['darkgrid','dark','whitegrid','white','ticks']
```

```python
sns.set_style(style[2])
sinplot()
```

```python
sns.set_style(style[0])
```

```python
sns.axes_style()
```

```python
sns.set_style(style[0],{'grid.color': 'pink'})
sinplot()
```

```python
# 恢复默认设置
sns.set()
sinplot()
```

## 8.2 plotting_context and set_context

```python
context = ['paper','notebook','talk','poster']
```

```python
sns.set_context(context[0])
sinplot()
```

```python
sns.set_context(context[1])  # 默认是notebook
sinplot()
```

```python
sns.set_context(context[2])
sinplot()
```

```python
sns.set_context(context[3])
sinplot()
```

```python
sns.set()
sinplot()
```

```python
# 显示当前context详细
sns.plotting_context()
```

```python
sns.set_context(context[1],rc = {'grid.linewidth': 5.0})  # 默认是notebook
sinplot()
```

```python
sns.set()
sinplot()
```

### 画板颜色大于6怎么操作？

```python
# 自定义RGB数值
pal = sns.color_palette([(0.8,0.1,0.2),(0.3,0.1,0.9)])
sns.palplot(pal)
```
